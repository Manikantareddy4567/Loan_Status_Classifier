```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('/content/loan_data.csv')
```

```python
data.dropna(inplace=True)
```

```python
data
```

```python
data.replace({"Loan_Status":{'N':0,'Y':1}},inplace=True)
```

```python
data.head()
```

```python
data['Dependents'].value_counts()
```

```python
data['Dependents'].replace(to_replace="3+",value=4,inplace=True)
data['Dependents'] = pd.to_numeric(data['Dependents'])
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encode=LabelEncoder()
my_list=['Gender','Married','Education','Self_Employed','Property_Area']
for x in my_list:
  data[x]=encode.fit_transform(data[x])
```

```python
data.head()
```

```python
data.drop('Loan_ID',axis=1,inplace=True)
```

```python
X=data.drop('Loan_Status',axis=1)
y=data['Loan_Status']
```

```python
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
model=SVC(C=0.1,kernel='linear',gamma='scale',degree=2)
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
model2=LogisticRegression()
model2.fit(X_train,y_train)
model2.score(X_test,y_test)
```

```python
final_model=RandomForestClassifier()
final_model.fit(X_train,y_train)
final_model.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
}
```

```python
grid_search=GridSearchCV(estimator=final_model,param_grid=param_grid,cv=5,n_jobs=-1,verbose=2)
```

```python
grid_search.fit(X_train,y_train)
```

```python
print(grid_search.best_params_)
print(grid_search.best_score_)
```

```python
final_model=RandomForestClassifier(n_estimators=300,min_samples_split=2,min_samples_leaf=1,max_depth=None)
final_model.fit(X_train,y_train)
```

```python
final_model.predict(X_test)
```

```python
def loan_status(input):
  prediction=final_model.predict(input)[0]
  if prediction==0:
    return 'Rejected'
  else:
    return 'Approved'
```

```python
my_input=list(map(float,input().split()))
```
